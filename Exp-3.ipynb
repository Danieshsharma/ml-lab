```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "auto3.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight", "engine-type",
           "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
           "peak-rpm","city-mpg","highway-mpg","price"
]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
df.head()
```

```python
import numpy as np
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#write your code below and press shift+enter to execute
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_horsepower = df['horsepower'].astype("float").mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan,"four", inplace=True)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd

# Example of loading data from a CSV file
df = pd.read_csv("auto3.csv")  # Replace with your actual file path

# Now perform your type conversion on the columns
df[["bore", "stroke"]] = df[["bore", "stroke"]].apply(pd.to_numeric, errors='coerce')
df[["normalized-losses"]] = df[["normalized-losses"]].apply(pd.to_numeric, errors='coerce').astype("Int64")
df[["price"]] = df[["price"]].apply(pd.to_numeric, errors='coerce')
df[["peak-rpm"]] = df[["peak-rpm"]].apply(pd.to_numeric, errors='coerce')

# Optionally, you can view the first few rows to confirm
print(df.head())
```
